```python
#b_iter = pd.read_csv('bookings.csv.bz2', sep='^', usecols=['off_time           ','dep_port','arr_port'], chunksize=1000000)
#forzamos a que sean todos los campos string con el dtype=str
#%%time
b_iter = pd.read_csv('bookings.csv.bz2', sep='^', chunksize=500000, low_memory=False, dtype=str)
bookings=pd.DataFrame()
for i, b_chunk in enumerate(b_iter):
    
    b=b_chunk.dropna()
    bookings = bookings.append(b)
    bookings.drop_duplicates(inplace=True)
    print((i+1)*500000, len(bookings))
bookings.to_csv('bookings_no_dup.csv', sep='^', index=False)
! bzip2 -f bookings_no_dup.csv
```
